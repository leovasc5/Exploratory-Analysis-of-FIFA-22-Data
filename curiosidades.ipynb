### Os 100 maiores potenciais.

```python
temp = df.sort_values(by=['potencial'], ascending=False)
pd.set_option('display.max_colwidth', 15)
temp = temp.loc[(temp['crescimento'])> 0]
temp = temp.loc[(temp['idade']) < 26]
temp[['nome', 'overall', 'idade', 'overall', 'potencial', 'crescimento', 'posicao_preferida', 'passe_euro', 'salario_euro', 'fim_contrato']].head(100)
```

### Nacionalidade dos 100 jogadores mais promisssores

```python
temp = temp[:100]
dados_paises = temp['nacionalidade'].value_counts()
dados_paises = dados_paises[:10]
dados_paises.head(15)
```

```python
plt.title("FIFA 22 - Nacionalidade dos 100 jogadores mais promissores (Barras)")
paises = ['Inglaterra', 'Espanha', 'França', 'Brasil', 'Itália', 'Portugal', 'Alemanha', 'Bélgica', 'Países Baixos', 'Argentina']
plt.bar(paises, dados_paises)
plt.show()
```

### OverAll Médio dos TOP 10 Países

```python
paises = ['England', 'Spain', 'France', 'Brazil', 'Italy', 'Portugal', 'Germany', 'Belgium', 'Netherlands', 'Argentina']
overall = []

for i in paises:
    overall.append(df.loc[(df['nacionalidade']) == i].overall.mean())
  
overall = np.round(overall, 1)

plt.title("FIFA 22 - OverAll Médio do TOP 10 Países (Barras)")
plt.plot(paises, overall, label='OverAll')
plt.show()
```
